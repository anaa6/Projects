```python
# Perbandingan Statistik Pokemon Legendary vs Non-Legendary
plt.figure(figsize=(10, 6))
sns.boxplot(x='Legendary', y='Total', data=df, palette='Set3')
plt.title('Perbandingan Total Pokemon Legendary vs Non-Legendary')
plt.xlabel('Legendary')
plt.ylabel('Total')
plt.show()
```

```python
# Menghitung jumlah total Pokemon dalam setiap generasi
jumlah_pokemon_per_generasi = df['Generation'].value_counts().sort_index()

# Menampilkan daftar jumlah Pokemon per generasi
print("Daftar Jumlah Total Pokemon dalam Setiap Generasi:")
for generasi, jumlah in jumlah_pokemon_per_generasi.items():
    print(f"Generasi {generasi}: {jumlah} Pokemon")
```

```python
# Analisis Tipe Pokemon Populer
type1_counts = df['Type 1'].value_counts()
type2_counts = df['Type 2'].value_counts()
all_types = pd.concat([type1_counts, type2_counts], axis=1).sum(axis=1)
popular_types = all_types.sort_values(ascending=False).head(10)

# Menghitung total Pokemon
total_pokemon = popular_types.sum()

# Menghitung persentase untuk setiap jenis tipe Pokemon populer
percentage_popular_types = (popular_types / total_pokemon) * 100

# Mencetak pesan dengan persentase untuk tipe Pokemon populer
most_popular_type = percentage_popular_types.idxmax()
print("Tipe Pokemon Paling Populer adalah tipe", most_popular_type)
print(f"Persentase Pokemon dengan tipe {most_popular_type}: {percentage_popular_types.loc[most_popular_type]:.1f}%")

# Visualisasi data Pokemon berdasarkan tipe dalam diagram lingkaran
plt.figure(figsize=(10, 8))
plt.pie(percentage_popular_types, labels=percentage_popular_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Pokemon Berdasarkan Tipe (Top 10)')
plt.axis('equal')  # Memastikan lingkaran menjadi persegi
plt.show()
```

```python
# Visualisasi distribusi kemunculan jenis Pokemon
plt.figure(figsize=(12, 6))
sns.barplot(x=popular_types.values, y=popular_types.index, palette='viridis')
plt.title('Distribusi Kemunculan Jenis PokÃ©mon')
plt.xlabel('Jumlah Kemunculan')
plt.ylabel('Jenis PokÃ©mon')

# Menambahkan label angka di ujung setiap bar
for index, value in enumerate(popular_types.values):
    plt.text(value, index, str(value), ha='left', va='center', fontsize=10, color='black')

plt.show()
```

```python
# Pokemon dengan Speed Tertinggi untuk Setiap Generasi
fastest_pokemon_index_per_generation = df.groupby('Generation')['Speed'].idxmax()

# Dapatkan informasi tentang Pokemon yang memiliki Speed tertinggi untuk setiap generasi
fastest_pokemon_per_generation = df.loc[fastest_pokemon_index_per_generation, ['Generation', 'Name']].reset_index(drop=True)

# Tampilkan hasil
print("\nPokemon dengan Speed Tertinggi untuk Setiap Generasi:")
print(fastest_pokemon_per_generation.to_string(index=False))
```

```python
# Pokemon dengan kecepatan terendah dan tertinggi
fastest_pokemon_index_global = df['Speed'].idxmax()
slowest_pokemon_index_global = df['Speed'].idxmin()

fastest_pokemon_global = df.loc[fastest_pokemon_index_global]
slowest_pokemon_global = df.loc[slowest_pokemon_index_global]

# Tampilkan pokemon yang paling cepat
print("Pokemon dengan kecepatan paling cepat dari seluruh pokemon:")
print("Nama:", fastest_pokemon_global['Name'])
print("Kecepatan:", fastest_pokemon_global['Speed'])
print("Generasi:", fastest_pokemon_global['Generation'])

# Tampilkan pokemon yang paling lambat
print("\nPokemon dengan kecepatan paling lambat dari seluruh pokemon:")
print("Nama:", slowest_pokemon_global['Name'])
print("Kecepatan:", slowest_pokemon_global['Speed'])
print("Generasi:", slowest_pokemon_global['Generation'])
```
